# Albedo Estimation
**Input:**

 - Intensity Image: $I$
 - Surface normal estimate (average from model): $n^{(0)}$
 
**Pre-calculated data**

 - $\sigma^2(s)$: estimated by generating images from Basel model under different lighting conditions and then estimating directions using the formula for $s^{(0)}$ below.
 - $\sigma^2_{ij}(n)$: estimated from Basel model
 - $\sigma^2_{ij}(\rho)$, $E(\rho_{ij})$, $E(\rho^2_{ij})$: estimated from albedo data created using Basel model

## Estimate light direction

$$
s^{(0)} = \left( \sum_{ij} n_{ij}^{(0)} {n_{ij}^{(0)}}^T \right)^{-1} \sum_{ij} I_{ij} n_{ij}^{(0)}
$$

which in matrix form is:

$$
s^{(0)} = (N N^T)^{-1} N I = N^+ I
$$

## Initial estimate of raw albedo

$$
\rho_{ij}^{(0)} = \frac{I_{ij}}{n_{ij}^{(0)} \cdot s^{(0)}}
$$

## Estimate non-stationary noise variance

$$
\sigma^2_{ij}(w) = \frac{\sigma^2_{ij}(n) + \sigma^2(s)}{(n_{ij}^{(0)} \cdot s^{(0)})^2}E(\rho^2_{ij})
$$

## Calculate the LMMSE estimate of the true unknown albedo

$$
\hat{\rho}_{ij} = (1 - \alpha_{ij})E(\rho_{ij}) + \alpha_{ij} \rho_{ij}^{(0)}
$$

$$
\alpha_{ij} = \frac{\sigma^2_{ij}(\rho)}{\sigma^2_{ij}(\rho) + \sigma^2_{ij}(w)}
$$

# Image normalisation
Create the improved albedo estimate (normalise the image to have $\rho = 1$)

$$
G_{ij} = \frac{I_{ij}}{\hat{\rho}_{ij}}
$$

**Pre-calculated data**

 - $\sigma^2_{ij}(H)$, $E(H_{ij})$: Generate randomly lit faces, estimate $s^{(0)}$ then calculate $H_{ij} = n_{ij} \cdot s^{(0)}$ using the known normals. Then estimate the ensemble values.

## Transform G to be relit by the estimate light direction

$$
\hat{H}_{ij} = (1 - \beta_{ij})E(H_{ij}) + \beta_{ij} G_{ij}
$$

$$
\beta_{ij} = \frac{\sigma^2_{ij}(H)}{\sigma^2_{ij}(H) + \sigma^2(s)}
$$

# Albedo Estimation in a multi-source scenario
Use the 9-light linear combination technique described by "Nine points of light: acquiring subspaces for for recognition under variable lighting".

Assume the following 9-light directions:

$$
\phi = \{0, 49, -68, 73, 77, -84, -84, 82, 50 \} \\
\theta = \{0, 17, 0, -18, 37, 47, -47, -56, -84 \}
$$

## Calculating the lighting coefficients

The image irradiance function now looks like 

$I = \sum_{k=1}^9 \gamma_k \rho (n \cdot s_k)$

and we can solve for $\gamma$ by performing the pseudoinverse $\gamma = W^+ I$ where $W^+$ is:

$$
W = \begin{pmatrix}
    max(n_1^{(0)} \cdot s_1, 0) & \ldots & max(n_1^{(0)} \cdot s_9, 0) \\
    max(n_2^{(0)} \cdot s_1, 0) & \ldots & max(n_2^{(0)} \cdot s_9, 0) \\
    \vdots & \ddots & \vdots \\
    max(n_N^{(0)} \cdot s_1, 0) & \ldots & max(n_N^{(0)} \cdot s_9, 0)
    \end{pmatrix}
$$

## Updating the LMMSE albedo estimate
As before, but update the noise variance as follows:

$$
\sigma^2(w) = \frac{\sum_{k=1}^9 \left[ \sigma^2(\gamma) E[(n \cdot s_k)^2] + \sigma^2(n) \gamma^2_k \right]}{\left( \sum_{k=1}^9 \gamma_k n^{(0)} \cdot s_k)^2 \right)} E(\rho^2)
$$

$$
E[(n \cdot s_k)^2] = (n^{(0)} \cdot s_k)^2 + \sigma^2(\gamma) \sigma^2(n)
$$
