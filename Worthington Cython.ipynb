```python
%%cython -a

cimport cython
import cython
import numpy as np
cimport numpy as np
from scipy.ndimage.filters import gaussian_filter
from libc.math cimport sin, cos, cosh

@cython.boundscheck(False)
cdef np.ndarray[np.float_t, ndim=2] axis_angle_to_rot(np.ndarray[np.float_t, ndim=1] vec, np.float_t angle):
    cdef np.ndarray[np.float_t, ndim=2] rm = np.zeros([3, 3])
    
    cdef np.float_t sA = sin(angle)
    cdef np.float_t cA = cos(angle)
    cdef np.float_t cAm = 1.0 - cA
    
    cdef np.float_t wx = vec[0]
    cdef np.float_t wy = vec[1]
    cdef np.float_t wz = vec[2]
    
    rm[0][0] = cA + wx*wx*cAm;     
    rm[1][0] = wz*sA + wx*wy*cAm
    rm[2][0] = -wy*sA + wx*wz*cAm
    
    rm[0][1] = wx*wy*cAm - wz*sA
    rm[1][1] = cA + wy*wy*cAm
    rm[2][1] = wx*sA + wy*wz*cAm
    
    rm[0][2] = wy*sA + wx*wz*cAm
    rm[1][2] = -wx*sA + wy*wz*cAm
    rm[2][2] = cA + wz*wz*cAm
    
    return rm

@cython.cdivision(True)
cdef double sech(np.float_t x):
    return 1.0 / cosh(x)

@cython.boundscheck(False)
cpdef np.ndarray[np.float_t, ndim=3] worthington_sfs(np.ndarray[np.float_t, ndim=2] image, 
                                                     np.ndarray[np.uint32_t, ndim=2] mask, 
                                                     np.ndarray[np.float_t, ndim=1] light, 
                                                     int n_iters=5):

    # Create the output needle map...
    cdef np.float_t sigmaZero = 10.0
    cdef int height = image.shape[0]
    cdef int width = image.shape[1]
    cdef np.ndarray[np.float_t, ndim=3] needle_map = np.zeros([height, width, 3])
    cdef np.ndarray[np.float_t, ndim=3] needle2 = np.zeros_like(needle_map)
    
    cdef np.float_t normR
    cdef np.float_t coneAng
    cdef np.ndarray[np.float_t, ndim=1] dx
    cdef np.ndarray[np.float_t, ndim=1] dy
    
    cdef np.float_t dx_s
    cdef np.float_t dy_s

    cdef np.float_t gradMag
    cdef np.ndarray[np.float_t, ndim=1] grad = np.zeros(3)
    cdef np.ndarray[np.float_t, ndim=1] gradPerp = np.zeros(3)
    cdef np.ndarray[np.float_t, ndim=2] rotMat
    
    cdef np.float_t sigma
    cdef np.float_t sigCc
    cdef np.float_t xc
    cdef np.float_t yx
    cdef np.float_t pis
    cdef np.float_t dxN
    cdef np.float_t dyN
    cdef np.float_t mult
    cdef np.ndarray[np.float_t, ndim=1] vec = np.zeros(3)
    cdef np.ndarray[np.float_t, ndim=1] axis = np.zeros(3)
    cdef np.float_t length
      
    # Make a blured version of the image, for making the gradiant from...
    cdef np.ndarray[np.float_t, ndim=2] smoothed_image = gaussian_filter(image, 1.44)
    
    # Prune from the mask the borders, as they don't have a valid region...
    for i in xrange(height):
        mask[i, 0] = 0
        needle_map[i, 0, 0] = 0.0
        needle_map[i, 0, 1] = 0.0
        needle_map[i, 0, 2] = 1.0
        mask[i, width - 1] = 0
        needle_map[i, width - 1, 0] = 0.0
        needle_map[i, width - 1, 1] = 0.0
        needle_map[i, width - 1, 2] = 1.0
    
    for i in xrange(width):
        mask[0, i] = 0
        needle_map[0, i, 0] = 0.0
        needle_map[0, i, 1] = 0.0
        needle_map[0, i, 2] = 1.0
        mask[height - 1, i] = 0
        needle_map[height - 1, i, 0] = 0.0
        needle_map[height - 1, i, 1] = 0.0
        needle_map[height - 1, i, 2] = 1.0

    # Initialise it...
    for x in xrange(width):
        for y in xrange(height):
            if mask[y, x]:
                if image[y, x] != 0.0:
                    # Assume divided by albedo already
                    normR = min(image[y, x], 1.0)
                    if not np.isfinite(normR):
                        normR = 1.0
                    coneAng = np.arccos(normR)
                    
                    dx_s = (smoothed_image[y+1,x-1] + 2.0 * smoothed_image[y+1,x] + smoothed_image[y+1,x+1]) - \
                         (smoothed_image[y-1,x-1] + 2.0 * smoothed_image[y-1,x] + smoothed_image[y-1,x+1])
                    dy_s = (smoothed_image[y-1,x+1] + 2.0 * smoothed_image[y,x+1] + smoothed_image[y+1,x+1]) - \
                         (smoothed_image[y-1,x-1] + 2.0 * smoothed_image[y,x-1] + smoothed_image[y+1,x-1])
                    grad[0] = dx_s
                    grad[1] = dy_s
                    grad[2] = 0.0
                    
                    # If magnitude of vector is 0.0
                    gradMag = np.sqrt(np.dot(grad, grad))
                    if gradMag == 0.0:
                        grad[0] = -light[0]
                        grad[1] = -light[1]
                        grad[2] = light[2]
                    else:  # Normalise
                        grad[0] /= gradMag
                        grad[1] /= gradMag
                        grad[2] /= gradMag
                    
                    gradPerp = np.cross(grad, light)
                    rotMat = axis_angle_to_rot(gradPerp, coneAng)
                    needle_map[y, x, :] = np.dot(rotMat, light)
                else:
                    mask[y, x] = 0.0
                    needle_map[y, x, 0] = 0.0
                    needle_map[y, x, 1] = 0.0
                    needle_map[y, x, 2] = 1.0

    # Iterate and smooth/correct the normals...
    for i in xrange(n_iters):
        # First pass to smooth and re-normalise
        for x in xrange(width):
            for y in xrange(height):
                if mask[y, x]:
                    sigma = 0.0
                    sicCc = 0.0
                    if mask[y - 1, x]:
                        dx = needle_map[y, x, :]
                        dx -= needle_map[y - 2, x, :]
                        dx *= 0.5
                        dy = needle_map[y - 1, x + 1, :]
                        dy -= needle_map[y - 1, x - 1, :]
                        dy *= 0.5
                        xc = 0.5 * (image[y, x] - image[y - 2, x]) - (np.dot(dx, light))
                        yc = 0.5 * (image[y - 1, x + 1] - image[y - 1, x - 1]) - (np.dot(dy, light))
                        sigma += np.exp(-(xc**2) - yc**2)
                        sigCc += 1
                        
                    if mask[y + 1, x]:
                        dx = needle_map[y + 2, x, :]
                        dx -= needle_map[y, x, :]
                        dx *= 0.5
                        dy = needle_map[y + 1, x + 1, :]
                        dy -= needle_map[y + 1, x - 1, :]
                        dy *= 0.5
                        xc = 0.5 * (image[y + 2, x] - image[y, x]) - (np.dot(dx, light))
                        yc = 0.5 * (image[y + 1, x + 1] - image[y + 1, x - 1]) - (np.dot(dy, light))
                        sigma += np.exp(-(xc**2) - yc**2)
                        sigCc += 1
                        
                    if mask[y, x - 1]:
                        dx = needle_map[y + 1, x - 1, :]
                        dx -= needle_map[y - 1, x - 1, :]
                        dx *= 0.5
                        dy = needle_map[y, x, :]
                        dy -= needle_map[y, x - 2, :]
                        dy *= 0.5
                        xc = 0.5 * (image[y + 1, x - 1] - image[y - 1, x - 1]) - (np.dot(dx, light))
                        yc = 0.5 * (image[y, x] - image[y, x - 2]) - (np.dot(dy, light))
                        sigma += np.exp(-(xc**2) - yc**2)
                        sigCc += 1
                        
                    if mask[y, x + 1]:
                        dx = needle_map[y + 1, x + 1, :]
                        dx -= needle_map[y - 1, x + 1, :]
                        dx *= 0.5
                        dy = needle_map[y, x + 2, :]
                        dy -= needle_map[y, x, :]
                        dy *= 0.5
                        xc = 0.5 * (image[y + 1, x + 1] - image[y - 1, x + 1]) - (np.dot(dx, light))
                        yc = 0.5 * (image[y, x + 2] - image[y, x]) - (np.dot(dy, light))
                        sigma += np.exp(-(xc**2) - yc**2)
                        sigCc += 1
                    
                    if sigCc:
                        sigma = (sigmaZero * sigma) / sigCc
                    sigma = max(sigma, 0.0001)
                    pis = np.pi / sigma
                    
                    dx = needle_map[y + 1, x, :]
                    dx -= needle_map[y - 1, x, :]
                    dx *= 0.5
                    dy = needle_map[y, x + 1, :]
                    dy -= needle_map[y, x - 1, :]
                    dy *= 0.5
                    
                    # Magnitudes of vectors
                    dxN = np.sqrt(np.dot(dx, dx))
                    dyN = np.sqrt(np.dot(dy, dy))
                    
                    if dxN != 0.0 and dyN != 0.0:
                        # Do the sum of neighbours in x part of the equation...
                        mult = np.tanh(pis * dxN) / dxN
                        vec = needle_map[y - 1, x, :]
                        vec += needle_map[y + 1, x, :]
                        vec *= mult
                        needle2[y, x, :] = vec
                        
                        # Do the big dx part of the equation...
                        mult = (pis * sech(pis * dxN)**2) / (dxN**2)
                        mult -= np.tanh(pis * dxN) / dxN**3
                        mult *= np.dot(dx.T, dx)
                        dx *= mult
                        needle2[y, x, :] += dx
                         
                        # Do the sum of neighbours in y part of the equation...
                        mult = np.tanh(pis * dyN) / dyN
                        vec = needle_map[y, x - 1, :]
                        vec += needle_map[y, x + 1, :]
                        vec *= mult
                        needle2[y, x, :] += vec
                       
                        # Do the big dy part of the equation...
                        mult = (pis * sech(pis * dyN)**2) / (dyN**2)
                        mult -= np.tanh(pis / dyN) / dyN**3
                        mult *= np.dot(dy.T, dy)
                        needle2[y, x, :] += dy

                    # Check the answer is good - if not fallback to DD1 from DD6...
                    # Get magnitude of needle2
                    length = np.sqrt(np.dot(needle2[y, x, :], needle2[y, x, :]))
                    if length == 0.0:  # Fallback to DD1
                        needle2[y, x, :] = np.asarray([0.0, 0.0, 0.0])
                        needle2[y, x, :] += needle2[y - 1, x, :]
                        needle2[y, x, :] += needle2[y + 1, x, :]
                        needle2[y, x, :] += needle2[y, x - 1, :]
                        needle2[y, x, :] += needle2[y, x + 1, :]
                        
                        # Magnitude of needle2
                        length = np.sqrt(np.dot(needle2[y, x, :], needle2[y, x, :]))
                        if length == 0.0:  # Fallback (unlikely)
                            needle2[y, x, :] = needle_map[y, x, :]
                            # Magnitude of needle2
                            length = np.sqrt(np.dot(needle2[y, x, :], needle2[y, x, :]))
                
                    # Renormalise, as numerical error adds up...
                    if length != 0.0:
                        needle2[y, x, 0] /= length
                        needle2[y, x, 1] /= length
                        needle2[y, x, 2] /= length
        
        # Enforce the cone constraint
        for x in xrange(width):
            for y in xrange(height):
                if mask[y, x]:
                    # Assumes already divided by Albedo
                    normR = min(image[y, x], 1.0)
                    if not np.isfinite(normR):
                        normR = 1.0
                    coneAng = np.arccos(normR)
                    
                    axis = np.cross(needle2[y, x, :], light)
                    # Sum of squares - same as magnitude?
                    length = np.sqrt(np.dot(needle2[y, x, :], needle2[y, x, :]))
                    if length == 0.0:
                        axis = np.asarray([1.0, 0.0, 0.0])
                    # Normalise vector
                    axis[0] /= length
                    axis[1] /= length
                    axis[2] /= length
                    
                    ang = np.arccos(np.clip(np.dot(needle_map[y, x, :], light), -1.0, 1.0)) - coneAng
                    
                    rotMat = axis_angle_to_rot(axis, ang)
                    needle_map[y, x, :] = np.dot(rotMat, needle2[y, x, :])
    
    # Renormalise the vectors before returning
    for x in xrange(width):
        for y in xrange(height):
            if mask[y, x]:
                length = np.sqrt(np.dot(needle_map[y, x, :], needle_map[y, x, :]))
                if length != 0.0:
                    needle_map[y, x, 0] /= length
                    needle_map[y, x, 1] /= length
                    needle_map[y, x, 2] /= length

    return needle_map
```

```python
import os.path
from photometric_stereo import photometric_stereo as ps
from pybug.image import MaskedNDImage, DepthImage, RGBImage
from pybug.io import auto_import
from surface_reconstruction import frankotchellappa, gradient_field_from_normals
from vector_utils import sph2cart, cart2sph
import numpy as np
import matplotlib.pyplot as plt


yale_b_path = '/vol/atlas/databases/yaleb/'
yale_b_ids = ['B01', 'B02', 'B03', 'B04', 'B05', 'B06', 'B07', 'B08', 'B09',
              'B10']

image_light_paths = [('yale{0}_P00A+000E+00.pgm', 0.0, 0.0),
                     ('yale{0}_P00A+000E-20.pgm', 0.0, -20.0),
                     ('yale{0}_P00A+020E-40.pgm', 20.0, -40.0),
                     ('yale{0}_P00A-020E-40.pgm', -20.0, -40.0),
                     ('yale{0}_P00A+035E+40.pgm', 35.0, 40.0),
                     ('yale{0}_P00A-035E+15.pgm', -35.0, 40.0)]

azimuths = np.asarray([np.radians(light[1]) for light in image_light_paths])
elevations = np.asarray([np.radians(light[2]) for light in image_light_paths])
lights = sph2cart(azimuths, elevations, np.ones(azimuths.shape[0]), theta_origin='z')
lights[:, [0, 1]] = lights[:, [1, 0]]
yale_id = 'B01'
images = [auto_import(os.path.join(yale_b_path, yale_id, light[0].format(yale_id)))[0] for light in image_light_paths]
```

```python
images_matrix = np.concatenate([im.pixels for im in images], axis=2)
photometric_stereo_image = MaskedNDImage(images_matrix)
photometric_stereo_image.landmarks = images[0].landmarks

photometric_stereo_image.crop_to_landmarks()
# photometric_stereo_image.constrain_mask_to_landmarks()

ground_truth_normals, ground_truth_albedo = ps(photometric_stereo_image,
                                               lights)
```

```python
test_image = photometric_stereo_image.pixels[:, :, 0]
mask = np.ones_like(test_image, dtype=np.uint32)
worthington_normals = worthington_sfs(test_image, mask, lights[0, :], n_iters=10)
```
